```python
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
import json
import numpy as np

pd.options.mode.chained_assignment = None  # default='warn'

#Pull in CSV paths

csvST = "csv data/Stress by State.csv"
stress_data = pd.read_csv(csvST)
stress_df = pd.DataFrame(stress_data)
```

```python
stress_clean = stress_df.dropna()
stress_filter = stress_clean.filter(['Event Category', 'Region'])
stress_group = stress_filter.groupby(['Region', 'Event Category']).agg({'Event Category': ['count']})
stress_final = stress_group.reset_index()
stress_final
```

```python
stress_dict = stress_final.to_dict('records')
stress_json = json.dumps(stress_dict, indent = 2)
print(stress_dict)
```
